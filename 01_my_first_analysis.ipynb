```python
# Cell 5: üé® CREATING THE FIRST CHART! 
```

```python
import matplotlib.pyplot as plt

# calculate top 10 currencies in dataset
top_currencies = df['symbol'].value_counts().head(10)

# bar-chart size
plt.figure(figsize=(10, 6))
plt.bar(top_currencies.index, top_currencies.values)

# label of the x-axis
plt.xlabel('Cryptocurrency Symbol')

# label of the y-axis
plt.ylabel('Count')

plt.title('Top 10 Currencies (JSON Dataset)')
plt.xticks(rotation=45)
plt.tight_layout()

# display/show the bar-chart
plt.show()

# printing/displaying the message
print("üéâ Chart created successfully!")
```
